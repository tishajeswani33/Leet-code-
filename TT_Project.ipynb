```python
import gradio as gr
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import io
from PIL import Image

def analyze_engagement():
    # Create synthetic engagement dataset
    np.random.seed(42)
    df = pd.DataFrame({
        "time_spent": np.random.randint(30, 300, 50),
        "messages_sent": np.random.randint(1, 40, 50),
        "questions_asked": np.random.randint(0, 15, 50)
    })

    # Plot
    plt.figure()
    plt.scatter(df['time_spent'], df['messages_sent'])
    plt.xlabel("Time Spent")
    plt.ylabel("Messages Sent")
    plt.title("Online Classroom Engagement Analysis")

    buf = io.BytesIO()
    plt.savefig(buf, format="png")
    plt.close()
    buf.seek(0)

    return Image.open(buf)

# Gradio UI
gr.Interface(
    fn=analyze_engagement,
    inputs=[],
    outputs=gr.Image(type="pil"),
    title="Online Classroom Engagement Analysis",
    description="Interactive visualization of online classroom engagement using synthetic data."
).launch()
```
